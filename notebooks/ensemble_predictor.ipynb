```python
import numpy as np
import joblib

# ----------------------------------------------------------
# LOAD MODELS
# ----------------------------------------------------------

def load_artifacts(name):
    model = joblib.load(f"models/{name}_model.pkl")
    scaler = joblib.load(f"models/{name}_scaler.pkl")
    imputer = joblib.load(f"models/{name}_imputer.pkl")
    return model, scaler, imputer

models = {
    "framingham": load_artifacts("framingham"),
    "cardio": load_artifacts("cardio"),
    "heart_failure": load_artifacts("heart_failure"),
    "diabetes": load_artifacts("diabetes"),
    "master": load_artifacts("master"),
}

# ----------------------------------------------------------
# WEIGHTS
# ----------------------------------------------------------

weights = {
    "framingham": 0.20,
    "cardio": 0.22,
    "heart_failure": 0.25,
    "diabetes": 0.18,
    "master": 0.15
}

# ----------------------------------------------------------
# FEATURE MAPPING PER MODEL
# ----------------------------------------------------------

def prepare_features(data, model_name):

    if model_name == "framingham":
        return [
            data["age"],
            data["sex"],
            data["bmi"],
            data["cholesterol"],
            data["sys_bp"],
            data["dia_bp"],
            data["glucose"],
            data["smoking"],
            data["hypertension"]
        ]

    elif model_name == "cardio":
        return [
            data["age"],
            data["sex"],
            data["bmi"],
            data["sys_bp"],
            data["dia_bp"],
            data["cholesterol"],
            data["smoking"]
        ]

    elif model_name == "heart_failure":
        return [
            data["age"],
            data["ejection_fraction"],
            data["serum_creatinine"],
            data["diabetes"],
            data["hypertension"]
        ]

    elif model_name == "diabetes":
        return [
            data["age"],
            data["bmi"],
            data["hypertension"],
            data["cholesterol"],
            data["smoking"],
            data["physical_activity"]
        ]

    elif model_name == "master":
        return [
            data["age"],
            data["sex"],
            data["bmi"],
            data["cholesterol"],
            data["diabetes"],
            data["smoking"],
            data["hypertension"],
            data["physical_activity"]
        ]

# ----------------------------------------------------------
# FINAL ENSEMBLE PREDICTION
# ----------------------------------------------------------

def predict_ensemble(data):

    final_score = 0

    for name in models:
        model, scaler, imputer = models[name]

        try:
            features = prepare_features(data, name)
            features = np.array([features])

            features = imputer.transform(features)
            features = scaler.transform(features)

            prob = model.predict_proba(features)[0][1]

            final_score += prob * weights[name]

        except:
            continue

    return round(final_score * 100, 2)
```
